### 02. Import Dataset

```python
df = pd.read_csv("data/processed/Telco_missing_values_handled.csv")
df.head()
```

### 03. Feature Binning

```python
def tenure_binning(tenure):
    if tenure <= 12:
        return 'Newer'
    elif tenure <= 48:
        return 'Medium'
    else:
        return 'Long-term'
```

```python
df['tenure_bins'] = df['tenure'].apply(tenure_binning)
del df['tenure']
```
